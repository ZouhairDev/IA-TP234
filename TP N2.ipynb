```python
#Importation des bibliothèques (numpy,matplotlib,sklearn)
#Pour manipuler dataset
import numpy as np
#Pour visualiser les données
import matplotlib.pyplot as plt
#Pour générer un nuage point(c-a-d simuler les données)
import sklearn
```

```python
#Importation du make_regression pour générer des ensembles de données de régression:
from sklearn.datasets import make_regression
```

```python
#fixer la graine du générateur de nombres aléatoires de numpy à 0:
np.random.seed(0)
```

```python
#générer un problème de régression aléatoire de 100 exemples et une seule variable:
x, y=make_regression(n_samples=100,n_features=1,noise=10)
```

```python
#Visualiser les données par méthode scatter :
plt.scatter(x,y)
```

```python
#Importation du SGRegressor pour créer et entraîner des modèles de régression linéaire:
from sklearn.linear_model import SGDRegressor
```

```python
#Définition du model sur 100 itérations et learning rate de 0,0001:
model = SGDRegressor(max_iter=100, eta0=0.0001)
```

```python
#Entrainer le model par méthode fit:
model.fit(x,y)
```

```python
#Calcul du coefficient par la fonction score:
print('Coeff Coefficient=',model.score(x,y))
```

```python
#Visualisation des données:
plt.scatter(x,y)
#Prédiction avec la fonction predict:
plt.plot(x,model.predict(x),c='red',lw=3)
```

```python
#Définition du model sur 100 0itérations et learning rate de 0,001
model = SGDRegressor(max_iter=1000, eta0=0.001)
```

```python
#Entrainer le model
model.fit(x,y)
```

```python
#Calcul du coefficient par la fonction score:
print('Coeff Coefficient=',model.score(x,y))
```

```python
#Visualisation des données
plt.scatter(x,y)
```

```python
#Prédiction avec la fonction predict:
plt.plot(x,model.predict(x),c='green',lw=3)
```

```python
#Définition du model sur 1000 itérations et learning rate de 0,001
model = SGDRegressor(max_iter=1000, eta0=0.001)
#Entrainer le model
model.fit(x,y)
#Calcul du coefficient par la fonction score:
print('Coeff Coefficient=',model.score(x,y))
```

```python
#Visualisation des données après entrainement du model et changement de nombre d'iteration et taux d'utilisation:
plt.scatter(x,y)
#Prédiction avec la fonction predict:
plt.plot(x,model.predict(x),c='green',lw=3)
```
